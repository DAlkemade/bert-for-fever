#Investigate predictions
This notebook allows the user to manually inspect evidence predictions for FEVER

```
from google.colab import drive
import json
import pprint
import pandas as pd
```

```
drive.mount('/content/drive')
pp = pprint.PrettyPrinter(indent=4)
```

```
with open('/content/drive/My Drive/Overig/dev_complete.sentences_id=k5t.p5.s5.jsonl', "r") as in_file:
# with open('/content/drive/My Drive/Overig/bert_sentence_predictions_only_classification=evidence_and_max_5.jsonl', "r") as in_file:
    instances = []
    for line in in_file:
        instances.append(json.loads(line))
print(len(instances))
```

```
for instance in instances[90:100]:
    if instance['verifiable'] == 'VERIFIABLE':
        all_evi = [[e[2], e[3]] for eg in instance["evidence"] for e in eg if e[3] is not None]
        pp.pprint(all_evi)
        pp.pprint(instance)
```

```
for instance in instances:
    if instance['id'] in [137334, 145446, 52175]:
        pp.pprint(instance)
```
